# Carregando os dados

```python
import pandas as pd

# Leitura do arquivo
df = pd.read_csv('titanic-data.csv', index_col='PassengerId')
df.head(10)
```

```python
# Analisando informações gerais do DataFrame.

df.info()
```

# Limpeza e preparação

Para não alterar o DataFrame original, farei uma cópia chamada dfAdjust.

```python
dfAdjust = df.copy()
dfAdjust.head()
```

## Removendo coluna Name

O nome dos passageiros não é necessário, visto que eles já possuem o identificador PassengerId como índice.

```python
dfAdjust = dfAdjust.drop('Name', axis=1)
dfAdjust.head()
```

## Separando Ticket em duas colunas

Alguns tickets possuem um tipo especial além dos números. Pode ser interessante analisar essas features separadamente.

```python
# Alocando os números dos tickets para uma coluna específica.

listTicketNumber = []
for item in dfAdjust['Ticket']:
    listTicketNumber.append(item.split()[-1])
dfAdjust['TicketNumber'] = listTicketNumber
dfAdjust.head()
```

```python
# Alocando os tipos dos tickets para uma coluna específica.

listTicketType = []
for item in dfAdjust['Ticket']:
    listTicketType.append(item.replace(item.split()[-1],''))
dfAdjust['TicketType'] = listTicketType
dfAdjust.head()
```

```python
# Identificando quantos tipos de ticket existem na base de dados.

for item in dfAdjust['TicketType'].unique():
    print(item)
```

Há muitos tipos diferentes de ticket, e talvez seja pouco prático analisar cada um isoladamente. Pode ser melhor separar entre tipo especificado e não-especificado.

```python
# Atribuindo um nome a valores vazios.

dfAdjust.loc[dfAdjust['TicketType']=='','TicketType'] = 'UNSPECIFIED'
dfAdjust.head()
```

```python
# Atribuindo 'SPECIFIED' a valores diferentes de 'UNSPECIFIED'.

dfAdjust.loc[dfAdjust['TicketType']!='UNSPECIFIED','TicketType'] = 'SPECIFIED'
dfAdjust.head()
```

```python
# Removendo a coluna Ticket e colocando TicketType e TicketNumber no seu lugar.

dfAdjust = dfAdjust.loc[:, ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch',
                            'TicketType', 'TicketNumber', 'Fare', 'Cabin', 'Embarked']]
dfAdjust.head()
```

## Criando faixas de idade

Lidar com idades de maneira puramente numérica pode atrapalhar as análises. Por isso, convém separar em menores de idade, adultos e idosos.

```python
dfAdjust.loc[:,'Age'] = df.Age
listAgeGroup = []

for item in dfAdjust['Age']:
    if item < 21:
        listAgeGroup.append('minor')
    elif item < 65:
        listAgeGroup.append('adult')
    elif item >= 65:
        listAgeGroup.append('elderly')
    else:
        listAgeGroup.append('unknown')

dfAdjust.loc[:,'Age'] = listAgeGroup
dfAdjust.head(10)
```

## Avaliando se convém remover Cabin

Há menos de 25% de objetos da base com cabines alocadas, então talvez não convenha mantê-la na base.

```python
print(dfAdjust[dfAdjust.Cabin.isnull()]['Survived'].value_counts())
print(dfAdjust[dfAdjust.Cabin.notnull()]['Survived'].value_counts())
```

Há uma prevalência de pessoas que sobreviveram tendo cabines alocadas, e prevalência de falecimentos sem cabines alocadas. Portanto, convém manter as cabines, mas os valores serão alterados para SPECIFIED e UNSPECIFIED.

```python
dfAdjust.loc[dfAdjust.Cabin.notnull(),'Cabin'] = 'SPECIFIED'
dfAdjust.loc[dfAdjust.Cabin.isnull(),'Cabin'] = 'UNSPECIFIED'

dfAdjust.head(10)
```

## Removendo coluna TicketNumber

Não faz sentido manter os números de tickets visto que eles não possuem características próprias em si. Apenas os tipos de tickets serão mantidos.

```python
dfAdjust = dfAdjust.drop('TicketNumber', axis=1)
dfAdjust.head(10)
```

## Conclusões

1. Passageiros de primeira classe tiveram uma chance claramente maior de sobreviver do que os demais.
2. Menores de idade tiveram uma chance significativamente maior de sobreviver do que os demais.
3. Passageiros do sexo feminino tiveram uma chance claramente maior de sobreviver do que os de sexo masculino.
4. Ter um tipo de ticket específico não fez diferença na chance de sobreviver.
5. Passageiros com uma cabine especificada tiveram uma chance claramente maior de sobreviver do que os sem cabines especificadas.
6. Passageiros que embarcaram em Cherbourg tiveram uma chance significativamente maior de sobreviver do que os demais.
